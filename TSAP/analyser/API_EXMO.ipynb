# POST ticker
https://api.exmo.com/v1.1/ticker

### Статистика по ценам и объему торгов по валютным парам
Statistics on prices and volume of trades by currency pairs. 

HEADERS: {Content-Type: application/x-www-form-urlencoded}

в результате получаем джейсон:
```
{
  "BTC_USD": {
    "buy_price": "589.06",
    "sell_price": "592",
    "last_trade": "591.221",
    "high": "602.082",
    "low": "584.51011695",
    "avg": "591.14698808",
    "vol": "167.59763535",
    "vol_curr": "99095.17162071",
    "updated": 1470250973
  }
}
```

значения, которые мы получаем по торговой паре:

1) "BTC_USD": - название валютной пары
2) "buy_price": - цена покупки
3) "sell_price": - цена продажи
4) "last_trade": - цена последней сделки
5) "high": - цена максимум за последние 24 часа
6) "low": - цена максимум за последние 24 часа
7) "avg": - цена средняя за последние 24 часа
8) "vol": - объём за 24 часа
9) "vol_curr": - объём валюты за 24 часа,
10) "updated": - обнавление дата как я понимаю


Скорее всего это история текущая информация по торговым парам  

````python
import requests

url = "https://api.exmo.com/v1.1/ticker"

payload={}
headers = {
  'Content-Type': 'application/x-www-form-urlencoded'
}

response = requests.request("POST", url, headers=headers, data=payload)

print(response.text)
````

### Написать функцию которая будет возвращать датафрейм, сделанный из этого джейсон ответа. 


````python
import pandas as pd
import requests

def ticker():
    url = "https://api.exmo.com/v1.1/ticker"

    payload={}
    headers = {
      'Content-Type': 'application/x-www-form-urlencoded'
    }

    response = requests.request("POST", url, headers=headers, data=payload)
    df = pd.read_json(response.text, orient='records').T
    
    curent_pairs_list = df.T.columns.values.tolist ()
    buy_price_list = df['buy_price'].tolist()
    sell_price_list = df['sell_price'].tolist()
    last_trade_list = df['last_trade'].tolist()
    high_list = df['high'].tolist()
    low_list = df['low'].tolist()
    avg_list = df['avg'].tolist()
    vol_list = df['vol'].tolist()
    vol_curr_list = df['vol_curr'].tolist()
    
    ticker_df = pd.DataFrame({'cur_pair' : curent_pairs_list,
                          'buy_price' : buy_price_list,
                          'sell_price': sell_price_list,
                          'last_trade': last_trade_list,
                          'high': high_list,
                          'low': low_list, 
                          'avg': avg_list, 
                          'vol': vol_list, 
                          'vol_curr': vol_curr_list 
                              
                         },
                           columns=['cur_pair',
                          'buy_price',
                          'sell_price',
                          'last_trade',
                          'high',
                          'low', 
                          'avg', 
                          'vol', 
                          'vol_curr' 
                                   ])

    
    return ticker_df
````

# Список сделок по валютным парам
- явно указываем торговую пару `payload='pair=BTC_USDT'`
- POST запрос

На выходе получаем джейсон:
- ключ - название торговой пары "BTC_USDT"
- значение - список значений - значения также являются словарями
```
"BTC_USDT":

{"trade_id":517030448,
"date":1695146118,
"type":"sell",
"quantity":"0.0015",
"price":"27171.32",
"amount":"40.75698"}
```

значения, которые мы получаем по торговой паре:

1) "trade_id": - id сделки
2) "date": - дата сделки (понять в каком виде)
3) "type": - тип сделки 
4) "quantity": - количество валюты
5) "price": - цена валюты
6) "amount": - общая сумма

Скорее всего это история торгов для конкретной торговой торговый пары. 

````python
import requests

url = "https://api.exmo.me/v1.1/trades"

payload='pair=SHIB_RUB'
headers = {
  'Content-Type': 'application/x-www-form-urlencoded'
}

response = requests.request("POST", url, headers=headers, data=payload)

print(response.text)
````

### Написать функцию, которая будет выдавать датафрейм, сделанный из полученного джейсона. 

````python
import pandas as pd
import requests

def trades():
    url = "https://api.exmo.me/v1.1/trades"

    payload='pair=BTC_RUB'
    headers = {
      'Content-Type': 'application/x-www-form-urlencoded'
    }

    response = requests.request("POST", url, headers=headers, data=payload)
    
    df = pd.read_json(response.text, orient='records')
    
    cur_pair = df.columns.tolist()[0]
    
    positions = range(0,len(df))
    
    curtent_pair = []
    trade_id = []
    date = []
    trd_type = []
    quantity = []
    price = []
    amount = [] 
    
    for el in positions:
        
        curtent_pair.append(cur_pair)
        trade_id.append(df.iloc[el][0]['trade_id'])
        date.append(df.iloc[el][0]['date'])
        trd_type.append(df.iloc[el][0]['type'])
        quantity.append(df.iloc[el][0]['quantity'])
        price.append(df.iloc[el][0]['price'])
        amount.append(df.iloc[el][0]['amount'])

    
    trades_df = pd.DataFrame({'cur_pair' : curtent_pair,
                              'trade_id': trade_id,
                              'date': date,
                              'type': trd_type,
                              'quantity': quantity,
                              'price': price,
                              'amount': amount
                         },
                           columns=['cur_pair',
                              'trade_id',
                              'date',
                              'type',
                              'quantity',
                              'price',
                              'amount'
                                   ])
    
    return trades_df
````

# POST order_book
https://api.exmo.me/v1.1/order_book

###  Книга текущих ордеров по валютной паре
The book of current orders on the currency pair

HEADERS:

- Content-Type: application/x-www-form-urlencoded
- Body: urlencoded
- pair: BTC_USDT,BTC_RUB (one or various currency pairs separated by commas)
- limit: 100 (the number of displayed positions (default: 100, max: 1000))

На выходе получаем джейсон:

```
{
  "BTC_USDT": {
    "ask_quantity": "3",
    "ask_amount": "500",
    "ask_top": "100",
    "bid_quantity": "1",
    "bid_amount": "99",
    "bid_top": "99",
    "ask": [
      [
        100,
        1,
        100
      ],
      [
        200,
        2,
        400
      ]
    ],
    "bid": [
      [
        99,
        1,
        99
      ]
    ]
  }
}
```
значения, которые мы получаем по торговой паре:

1) "BTC_USDT": - валютная пара
2) "ask_quantity": - количество спроса
3) "ask_amount": - сумма спроса
4) "ask_top": - топовый спрос
5) "bid_quantity": - количество предложения
6) "bid_amount": - сумма предложений
7) "bid_top": - топовое предложение 
8) "ask": - список ордеров спроса : количество покупаемой валюты, цена, сумма 
9) "bid": - список ордеров предложений: количество продоваемой валюты, цена, сумма 

Это скорее всего книга предложений, которые открыты на бирже по данной валютной паре. 

````python
import requests

url = "https://api.exmo.me/v1.1/order_book"

payload='pair=BTC_USDT&limit=100'
headers = {
  'Content-Type': 'application/x-www-form-urlencoded'
}

response = requests.request("POST", url, headers=headers, data=payload)

print(response.text)
````

### Написать функцию, которая будет возвращать:
- датафрейм по полученному джейсону
- датафрейм по спросу
- датафрейм по предложению
- вычисляемая цена

````python
import pandas as pd
import requests

def order_book():
    url = "https://api.exmo.me/v1.1/order_book"

    payload='pair=BTC_RUB&limit=200'
    headers = {
      'Content-Type': 'application/x-www-form-urlencoded'
    }

    response = requests.request("POST", url, headers=headers, data=payload)
    df = pd.read_json(response.text, orient='records')
    
    cur_pair = df.columns.tolist()[0]
    
    curtent_pair = []
    ask = []
    ask_amount = []
    ask_quantity = []
    ask_top = []
    bid = []
    bid_amount = []
    bid_quantity = []
    bid_top = []
    
    curtent_pair.append(cur_pair)
    ask.append(df.T['ask'].tolist()[0])
    ask_amount.append(df.T['ask_amount'].tolist()[0])
    ask_quantity.append(df.T['ask_quantity'].tolist()[0])
    ask_top.append(df.T['ask_top'].tolist()[0])
    bid.append(df.T['bid'].tolist()[0])
    bid_amount.append(df.T['bid_amount'].tolist()[0])
    bid_quantity.append(df.T['bid_quantity'].tolist()[0])
    bid_top.append(df.T['bid_top'].tolist()[0])
    
    order_book_df = pd.DataFrame({'cur_pair' : curtent_pair,
                                  'ask': ask,
                                  'ask_amount': ask_amount,
                                  'ask_quantity':  ask_quantity,
                                  'ask_top':  ask_top,
                                  'bid':  bid,
                                  'bid_amount':  bid_amount,
                                  'bid_quantity':  bid_quantity,
                                  'bid_top':  bid_top
                         },
                           columns=['cur_pair',
                                  'ask',
                                  'ask_amount',
                                  'ask_quantity',
                                  'ask_top',
                                  'bid',
                                  'bid_amount',
                                  'bid_quantity',
                                  'bid_top'
                                   ])
      
    cur_pair_ask = []
    price_ask = []
    amount_ask = []
    quantity_ask = []
    
    for el in range(0, len(order_book_df['ask'][0])):
        cur_pair_ask.append(cur_pair)
        price_ask.append(float(order_book_df['ask'][0][el][0]))
        amount_ask.append(float(order_book_df['ask'][0][el][1]))
        quantity_ask.append(float(order_book_df['ask'][0][el][2]))

    ask_df = pd.DataFrame({'cur_pair': cur_pair_ask,
                       'price_ask': price_ask,
                       'amount_ask': amount_ask,
                       'quantity_ask': quantity_ask
                      },
                     columns=['cur_pair',
                       'price_ask',
                       'amount_ask',
                       'quantity_ask'])
    
    cur_pair_bid = []
    price_bid = []
    amount_bid = []
    quantity_bid = []

    for el in range(0, len(order_book_df['bid'][0])):
        cur_pair_bid.append(cur_pair)
        price_bid.append(float(order_book_df['bid'][0][el][0]))
        amount_bid.append(float(order_book_df['bid'][0][el][1]))
        quantity_bid.append(float(order_book_df['bid'][0][el][2]))

    bid_df = pd.DataFrame({'cur_pair': cur_pair_bid,
                       'price_bid': price_bid,
                       'amount_bid': amount_bid,
                       'quantity_bid': quantity_bid
                      },
                     columns=['cur_pair',
                       'price_bid',
                       'amount_bid',
                       'quantity_bid'])
   
    price_cur = (price_ask[0] + price_bid[0]) / 2
                                
    return order_book_df, ask_df, bid_df, price_cur
````

`HEADERS:` {Content-Type: application/x-www-form-urlencoded}

На выходе получаем джейсон:
```
{
  "EXM_ETH": {
    "min_quantity": "1",
    "max_quantity": "1000",
    "min_price": "1",
    "max_price": "1000",
    "max_amount": "1000",
    "min_amount": "1",
    "price_precision": 8,
    "commission_taker_percent": "0.2",
```
значения, которые мы получаем по торговой паре:

- "EXM_ETH": - Валютная пара
- "min_quantity": - минимальное количество
- "max_quantity": - максимальное количество
- "min_price": - минимальная цена
- "max_price": - максимальная цена
- "min_amount": - минимальная сумма
- "price_precision": - точность цены
- "commission_taker_percent": комиссия

Скорее всего это настройки торговой торговый пары. 

````python
import requests

url = "https://api.exmo.com/v1.1/pair_settings"

payload={}
headers = {
  'Content-Type': 'application/x-www-form-urlencoded'
}

response = requests.request("POST", url, headers=headers, data=payload)

print(response.text)
````

### Написать функцию которая будет возвращать датафрейм, сделанный из этого джейсон ответа. 

````python
import pandas as pd
import requests

def pair_settings():
    url = "https://api.exmo.com/v1.1/pair_settings"

    payload={}
    headers = {
      'Content-Type': 'application/x-www-form-urlencoded'
    }

    response = requests.request("POST", url, headers=headers, data=payload)
    
    df = pd.read_json(response.text, orient='records').T
    
    curent_pairs_list = df.T.columns.values.tolist ()
    min_quantity_list = df['min_quantity'].tolist()
    max_quantity_list = df['max_quantity'].tolist()
    min_price_list = df['min_price'].tolist()
    max_price_list = df['max_price'].tolist()
    min_amount_list = df['min_amount'].tolist()
    price_precision_list = df['price_precision'].tolist()
    commission_taker_percen_list = df['commission_taker_percent'].tolist()
    
    pair_settings_df = pd.DataFrame({'cur_pair' : curent_pairs_list,
                          'min_quantity' : min_quantity_list,
                          'max_quantity': max_quantity_list,
                          'min_price': min_price_list,
                          'max_price': max_price_list,
                          'min_amount': min_amount_list, 
                          'price_precision': price_precision_list, 
                          'commission_taker_percent': commission_taker_percen_list, 

                         },
                           columns=['cur_pair',
                          'min_quantity' ,
                          'max_quantity',
                          'min_price',
                          'max_price',
                          'min_amount', 
                          'price_precision', 
                          'commission_taker_percent'
                                   ])

    
    return pair_settings_df
````

# POST required_amount
https://api.exmo.com/v1.1/required_amount

Calculating the sum of buying a certain amount of currency for the particular currency pair
### Расчет суммы покупки определенного количества валюты для конкретной валютной пары

- `HEADERS`: {Content-Type: application/x-www-form-urlencoded}
- `Body`: urlencoded
- `pair`: BTC_USD currency pair
- `quantity`: 11 quantity to buy

На выходе получаем джейсон:
```
{
  "quantity": 3,
  "amount": 5,
  "avg_price": 3.66666666
}
```

````python
import requests

url = "https://api.exmo.com/v1.1/required_amount"

payload='pair=BTC_USD&quantity=11'
headers = {
  'Content-Type': 'application/x-www-form-urlencoded'
}

response = requests.request("POST", url, headers=headers, data=payload)

print(response.text)
````
